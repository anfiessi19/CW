```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
```

#### Зареєструватись на платформі kaggle і завантажити дані з https://www.kaggle.com/c/titanic/data?select=test.csv (також файл з даними прикріплено у Google Classroom)

```python
df = pd.read_csv("train.csv", index_col="PassengerId")
df.head()
```

* Survived - вижила людина чи ні, цільова змінна
* Pclass - клас, в якому їхала людина
* Name - ім'я людини
* Sex - стать
* Age - вік
* SibSp - скільки було на борту братів, сестер, подружжя на борту Титаніка
* Parch - кількість дітей та батьків на борту Титаніка
* Ticket - номер квитка
* Fare - вартість квитка
* Cabin - номер каюти
* Embarked - порт посадки (C = Шербур, Q = Квінстаун, S = Саутгемптон)

Залишіть у вибірці три ознаки: клас пасажира (Pclass), вік пасажира (Age) та його стать (Sex).

```python
X = df.loc[:, ["Pclass", "Age", "Sex"]]
X.head()
```

Зверніть увагу, що ознака Sex має рядкові значення.

```python
X["Sex"] = X["Sex"].map({"male": 0, "female": 1})
```

Виділіть цільову змінну - вона записана в стовпці Survived.

```python
y = df["Survived"]
```

У даних є пропущені значення - наприклад, для деяких пасажирів невідомий їхній вік.

Такі записи під час читання в pandas приймають значення nan. Знайдіть усі об'єкти, які мають пропущені ознаки, та видаліть їх з вибірки.

```python
X.info()
```

```python
X = X.dropna()
y = y[X.index]
```

Виділимо 70% вибірки (X_train, y_train) під навчання та 30% будуть відкладеною вибіркою (X_test, y_test). Відкладена вибірка ніяк не братиме участі в налаштуванні параметрів моделей, на ній ми наприкінці оцінимо якість отриманої моделі.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

#### Навчимо метод **дерево рішень** з параметрами по замовчуванню

```python
tree_1 = DecisionTreeClassifier(random_state=241)
tree_1.fit(X_train, y_train)
```

Якість прогнозів перевірятимемо за допомогою простої метрики – частки правильних відповідей.

```python
pred_1 = tree_1.predict(X_test)
accuracy_score(y_test, pred_1)
```

Основні параметри класу sklearn.tree.DecisionTreeClassifier:

 - max_depth – максимальна глибина дерева
 - max_features - максимальна кількість ознак, за якими шукається краще розбиття в дереві (це потрібно тому, що при великій кількості ознак буде "затратно" шукати краще (за критерієм типу приросту інформації) розбиття серед усіх ознак)
 - min_samples_leaf – мінімальна кількість об'єктів у листі. Цей параметр має зрозумілу інтерпретацію: скажімо, якщо він дорівнює 5, то дерево породжуватиме тільки ті класифікуючі правила, які вірні як мінімум для 5 об'єктів.


```python
tree_2 = DecisionTreeClassifier(min_samples_leaf = 4, max_depth = 5)
tree_2.fit(X_train, y_train)
```

```python
pred_2 = tree_2.predict(X_test)
accuracy_score(y_test, pred_2)
```

```python
plt.figure(figsize=(8,8))
plot_tree(tree_2, filled=True)
plt.show()
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_2, out_file='my_tree.dot', filled=True)
```

#### Навчимо метод **KNN** з параметрами по замовчуванню

```python
knn_1 = KNeighborsClassifier()
knn_1.fit(X_train, y_train)
```

```python
pred_3 = knn_1.predict(X_test)
accuracy_score(y_test, pred_3)
```

Основні параметри класу sklearn.neighbors.KNeighborsClassifier:

 - n_neighbors: кількість сусідів
 - metric: "minkowski", "manhattan", "euclidean", "chebyshev" та інші
 - weights: "uniform" (всі ваги рівні), "distance" (вага обернено пропорційна відстані до тестового прикладу) або інша певна користувачем функція
 - algorithm (опціонально): "brute", "ball_tree", "KD_tree", або "auto". У першому випадку найближчі сусіди кожного тестового прикладу рахуються перебором навчальної вибірки. У другому та третьому – відстані між об'єктами зберігаються у дереві, що прискорює знаходження найближчих сусідів. У разі вказівки параметра "auto" підходящий спосіб знаходження сусідів буде обраний автоматично на основі навчальної вибірки.
 - leaf_size (опціонально): поріг перемикання на повний перебір у разі вибору BallTree або KDTree для знаходження сусідів


```python
knn_2 = KNeighborsClassifier(n_neighbors = 3, metric = "euclidean", weights = "distance")
knn_2.fit(X_train, y_train)
pred_4 = knn_2.predict(X_test)
accuracy_score(y_test, pred_4)
```
