# Price Prediction Of Diamonds: Regression

### About Dataset

This classic dataset contains the prices and other attributes of almost 54,000 diamonds. There are 10 attributes included in the dataset including the target ie. price.

Feature description:

* **Carat :** Carat weight of the Diamond. (вага діаманта в каратах)
* **Cut :** Describe cut quality of the diamond. (якість огранювання діаманта)
> * Quality in increasing order Fair, Good, Very Good, Premium, Ideal .
* **Color :** Color of the Diamond.
> * With D being the best and J the worst.
* **Clarity :** Diamond Clarity refers to the absence of the Inclusions and Blemishes. (відсутність включень і плям)
> * (In order from Best to Worst, FL = flawless, I3= level 3 inclusions) FL, IF, VVS1, VVS2, VS1, VS2, SI1, SI2, I1, I2, I3
* **Depth :** The Height of a Diamond, measured from the Culet to the table, divided by its average Girdle Diameter.
* **Table :** The Width of the Diamond's Table expressed as a Percentage of its Average Diameter. (відношення ширини верхньої грані діаманта до його максимальної ширини, у відсотках)
* **Price :** the Price of the Diamond.
* **X :** Length of the Diamond in mm.
* **Y :** Width of the Diamond in mm.
* **Z :** Height of the Diamond in mm.

Qualitative Features (Categorical) : Cut, Color, Clarity.

Quantitative Features (Numerical) : Carat, Depth , Table , Price , X , Y, Z.

### Price is the Target Variable.
price in US dollars (326 -- 18,823)This is the target column containing tags for the features.

## Data_cleaning

**First**, we will drop the unnamed column

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

**Second**, we noticed that the dimensions columns (x,y,z) has zero values which is wrong values and will affect model so we will
drop these columns from dataset

```python
df.drop(df[df['x']==0].index,inplace=True)
df.drop(df[df['y']==0].index,inplace=True)
df.drop(df[df['z']==0].index,inplace=True)
```

**Third**, we will remove the outliers in the following columns (Depth,Table,X,Y,Z)

```python
df=df[(df['depth']>49) & (df['depth']<75)]
df=df[(df['table']>49) & (df['table']<70)]
df=df[(df['x']>2.8) & (df['x']<20)]
df=df[(df['y']>1) & (df['y']<15)]
df=df[(df['z']>1) & (df['z']<7)]
```

```python
df.shape
```

**Fourth**, we will convert the categorical columns with the label encoder function to include them in model to give me a better accuracy using label encoder function since it is an ordinal column which means that it has rank and order

```python
label_encoder=LabelEncoder()
```

```python
df['cut']=label_encoder.fit_transform(df['cut'])
df['color']=label_encoder.fit_transform(df['color'])
df['clarity']=label_encoder.fit_transform(df['clarity'])
```
