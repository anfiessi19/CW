# **Бібліотека numpy**

Пакет `numpy` надає $n$-мірні однорідні масиви (всі елементи одного типу); у них не можна вставити або видалити елемент у довільному місці. У `numpy` реалізовано багато операцій над масивами.

## 1. Одновимірні масиви

#### 1.1 Типи масивів, атрибути

Можна перетворити список на масив.

```python
a = np.array([0, 2, 1])
a, type(a)
```

`print` друкує масиви у зручній формі.

```python
print(a)
```

Клас `ndarray` має багато методів.

Цей масив одновимірний.

```python
a.ndim
```

У $n$-вимірному випадку повертається кортеж розмірів за кожною координатою.

```python
a.shape
```

`size` - це повна кількість елементів у масиві; `len` - розмір за першою координатою (в 1-мірному випадку це те саме).

```python
len(a), a.size
```

#### 1.2 Індексація

Індексувати масив можна звичайним чином.

```python
a[1]
```

Масиви можна змінювати.

```python
a[1] = 3
print(a)
```

Масиви, зрозуміло, можна використовувати в циклах. Але при цьому втрачається головна перевага `numpy` – швидкодія. Завжди, коли це можливо, краще використовувати операції над масивами, як єдиним цілим.

```python
for i in a:
    print(i)
```

#### 1.3 Створення масивів

Масиви, заповнені нулями чи одиницями. Часто краще спочатку створити такий масив, а потім надавати значення його елементам.

```python
a = np.zeros(3)
b = np.ones(3, dtype=np.int64)
print(a)
print(b)
```

Функція `arange` подібна до `range`. Аргументи можуть бути з плаваючою точкою. Слід уникати ситуацій, коли *(кінець-початок)/крок* - ціле число, оскільки у разі включення останнього елемента залежить від помилок округлення. Краще щоб кінець діапазону був десь посередині кроку.

```python
a = np.arange(0, 9, 2)
print(a)
```

```python
b = np.arange(0., 9, 2)
print(b)
```

Послідовності чисел із постійним кроком можна також створювати функцією `linspace`. Початок та кінець діапазону включаються; останній аргумент – кількість точок.

```python
a = np.linspace(0, 8, 5)
print(a)
```

## 5. Лінійна алгебра

У підбібліотеці numpy.linalg є корисні функції для роботи з матрицями. Наприклад:

- `linalg.cholesky` - розкладя Холецького,
- `linalg.qr` - QR-розклад,
- `linalg.svd` - сингулярне матричне розкладання,
- `linalg.eig` - пошук власних значень та векторів,
- `linalg.det` - визначник матриці,
- `linalg.norm` - норма матриці,
- `linalg.inv` - обернена матриця,
- `linalg.solve` - розв'язання системи лінійних рівнянь, заданої в матричному вигляді.

Наприклад, знайдемо визначник
$$\left|
\begin{array}{cc}
1 & 2 \\
-2 & 3
\end{array}
\right|.$$

```python
a = np.array([[0, 1], [-2, 3]])
```

```python
np.linalg.det(a)
```

Обернена матриця.

```python
a1 = np.linalg.inv(a)
print(a1)
```

Розв'язати систему рівнянь
$$
\left\{
\begin{array}{ll}
3x + y &= 2, \\
 x - y &= 3.
\end{array}
\right.
$$

```python
A = np.array([[3, 1], [1, -1]])
b = np.array([2, 3])
x = np.linalg.solve(A, b)
print(x)
```

## 6. Продуктивність numpy

Подивимося на простий приклад --- сума перших $10^8$ чисел.

```python
%%time

sum_value = 0
for i in range(10 ** 8):
    sum_value += i
print(sum_value)
```

Код із використанням функцій бібліотеки numpy

```python
%%time
sum_value = 0
sum_value = np.arange(10 ** 8).sum(dtype=np.int64)
print(sum_value)
```

Обчислюємо синус кожного елемента у матриці $5000\times 5000$ у вигляді списку списків

```python
from math import sin

size = 5000
python_list = [[size * i + j for j in range(size)] for i in range(size)]
```

```python
%%time
for i in range(size):
    for j in range(size):
        python_list[i][j] = sin(python_list[i][j])
```

У вигляді масиву numpy:

```python
numpy_array = np.arange(size * size).reshape((size, size))
```

```python
%%time
numpy_array = np.sin(numpy_array)
```
